```python
#SINGLE CALL FOR ARTIST DATAFRAME
#Import needed libraries
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

#Authenticate with Spotify API
client_id ='b5762be2d23d43b189262e8c53a6078c'
client_secret = 'a7e8d83649314303817889927c027728'
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)
def ArtistDataFrame(artist_name):
    #Get a given Artist's Spotify ID
    get_id_hh = sp.search(artist_name, type='artist').get('artists').get('items')[0].get('id')
    get_id_hh

    #Pull top tracks for artist
    tracks_hh=sp.artist_top_tracks(get_id_hh)
    tracks_hh

    #Create DataFrame with top tracks for artist
    number_of_tracks = len(tracks_hh.get('tracks'))
    artist_df = pd.DataFrame()
    for x in range(0,number_of_tracks):
        track_title = tracks_hh.get('tracks')[x].get('name')
        track_id = tracks_hh.get('tracks')[x].get('id')
        track_artist = tracks_hh.get('tracks')[1].get('artists')[0].get('name')
        artist_id= tracks_hh.get('tracks')[x].get('artists')[0].get('id')
        track_album = tracks_hh.get('tracks')[x].get('album').get('name')
        album_id= tracks_hh.get('tracks')[x].get('album').get('id')
        track_release_date = tracks_hh.get('tracks')[x].get('album').get('release_date')
        album_type =  tracks_hh.get('tracks')[x].get('album').get('album_type')
        track_popularity = tracks_hh.get('tracks')[x].get('popularity')
        columns = {'track_title':[track_title],
                    'track_id':[track_id],
                    'track_artist':[track_artist],
                    'artist_id':[artist_id],
                    'track_album':[track_album],
                    'track_release_date':[track_release_date],
                    'album_type':[album_type],
                    'track_popularity':[track_popularity]

                          }
        temp_df=pd.DataFrame.from_dict(columns)
        artist_df= pd.concat([artist_df, temp_df])
    artist_df

    #Pull Metadata for Artist top tracks
    songmeta_temp = {}
    meta_df= pd.DataFrame()
    for i,row in artist_df.iterrows():
        #print(i)
        try:
            song_features = sp.audio_features(row['track_id'])[0]
            artist_info = sp.artist(row['artist_id'])
            artist_followers = artist_info.get('followers').get('total')
            artist_genres = artist_info.get('genres')
            artist_popularity = artist_info.get('popularity')
            if song_features == None or artist_info == None:
                continue
            else:
                for k,v in song_features.items():
                    songmeta_temp[k]=[v]
                songmeta_temp['artist_followers']=[artist_followers]
                songmeta_temp['artist_genres']=[artist_genres]
                songmeta_temp['artist_popularity']=[artist_popularity]
                song_features_temp_df = pd.DataFrame.from_dict(songmeta_temp)
                meta_df = pd.concat([meta_df, song_features_temp_df])
        except:
            continue

    #Join MetaData to Artist Dataframe
    artist_merged = pd.merge(artist_df, meta_df, how = 'left', left_on='track_id', right_on='id' )
    artist_merged.drop(['id', 'uri', 'track_href'], axis = 1, inplace=True)
    return artist_merged
```
